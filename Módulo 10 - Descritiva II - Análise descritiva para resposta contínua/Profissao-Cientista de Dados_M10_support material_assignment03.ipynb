1) Construa um gráfico de barras para avaliar a distribuição das variáveis qualitativas no tempo. Você acha que houve alguma alteração importante a ser investigada?

````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn

df = pd.read_csv('previsao_de_renda.csv')
df.dropna(inplace=True)
df
````

````python
def make_graph(hue: str):
    fig = plt.gcf()
    fig.set_size_inches( 10, 5)


    g = sns.countplot(data=df, x='data_ref', hue=hue)
    g = plt.xticks(rotation=90)
    return g
````

````python
make_graph('tipo_renda')
````

````python
make_graph('posse_de_veiculo')
````

````python
make_graph('educacao')
````

````python
make_graph('posse_de_imovel')
````

````python
make_graph('tipo_residencia')
````

Houve uma alteração significativa da variável 'educacao', em relação ao número de pessoas com nível superior completo e nível secundário, a parftir de setembro de 2015.

2) Para cada variável qualitativa, construa um gráfico mostrando as médias de cada nível da variável no tempo e avalie a estabilidade. Houve alguma alteração importante? Você acha que se pode utilizar essas variáveis em uma modelagem?

````python
def make_graph_media(hue: str):
    fig = plt.gcf()
    fig.set_size_inches( 10, 5)


    g = sns.pointplot(data=df, x='data_ref', y='renda', hue=hue, dodge=True)
    g = plt.xticks(rotation=90)
    return g
````

````python
make_graph_media('educacao')
````

````python
make_graph_media('tipo_renda')
````

````python
make_graph_media('posse_de_veiculo')
````

````python
make_graph_media('posse_de_imovel')
````

````python
make_graph_media('tipo_residencia')
````

As variáveis 'posse de veículo', 'posse de imóvel' e 'tipo de residência' apresentam comportamento estável, não apresentando mudanças significativas ao longo do tempo. Poderiam ser usadas para uma modelagem.
