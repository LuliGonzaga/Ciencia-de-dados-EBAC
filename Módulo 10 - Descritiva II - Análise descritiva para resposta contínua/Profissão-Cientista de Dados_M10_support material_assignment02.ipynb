````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn

df = pd.read_csv('previsao_de_renda.csv')
df.dropna(inplace=True)
df
````

````python
ax = sns.pointplot(x="posse_de_imovel", y="renda",
                   data=df, dodge=True, ci = 95)
````

````python
ax = sns.pointplot(x="posse_de_veiculo", y="renda",
                   data=df, dodge=True, ci = 95)
````

As variáveis "posse de veículo" e "posse de imóvel" são boas preditoras de renda pois, segundo a análise dos gráficos, em ambos os casos a renda média apresenta uma diferença significativa em relação a quem tem ou não um desses itens, sendo maior no caso verdadeiro.
