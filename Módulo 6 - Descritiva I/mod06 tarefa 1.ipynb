```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
sinasc_RO = pd.read_csv('sinasc_RO_2019.csv')
sinasc_RO.head(3)
```

####  2) Monte um gráficos de barras com as variáveis APGAR1, APGAR5, SEXO, GESTACAO, ESTCIVMAE.

```python
sns.countplot(x='APGAR1', data=sinasc_RO)
```

```python
sns.countplot(x='APGAR5', data=sinasc_RO)
```

```python
sns.countplot(x='SEXO', data=sinasc_RO)
```

```python
sns.countplot(x='GESTACAO', data=sinasc_RO).tick_params(axis="x", rotation = 30)
```

```python
sns.countplot(x='ESTCIVMAE', data=sinasc_RO)
```

#### 3) Monte um gráfico de barras mostrando o APGAR5 médio por CONSULTAS

```python
sns.barplot(x='CONSULTAS', y='APGAR5' , data=sinasc_RO)
```

####  4) Monte um painel com 4 gráficos mostrando as frequencias de SEXO, GESTACAO, ESTCIVMAE e CONSULTAS

```python
plt.rc('figure', figsize=(10, 8))
fig, axes = plt.subplots(2, 2)

sns.countplot(ax = axes[0, 0], x='SEXO', data=sinasc_RO)
sns.countplot(ax = axes[0, 1], x='GESTACAO', data=sinasc_RO).tick_params(axis="x", rotation = 30)
sns.countplot(ax = axes[1, 0], x='ESTCIVMAE', data=sinasc_RO)
sns.countplot(ax = axes[1, 1], x='CONSULTAS', data=sinasc_RO)

fig.tight_layout()
```
