### 1. Idade media das mães e dos pais por município (coluna munResNome)


```python
sinasc_1 = sinasc_raw[['munResNome','IDADEMAE','IDADEPAI']]
sinasc_1.head()
```

### 3. Qual o municipio que nasceu menos bebe em 2019?
    - qual a idade media, maxima, minima das maes nesse municipio?
    - qual a idade media, maxima, minima dos pais nesse municipio?

```python
sinasc_raw['munResNome'].value_counts()
```

O município de Castanheiras teve o menor número de nascimentos de 2019 (32)

```python
sinasc_3 = sinasc_raw.loc[(sinasc_raw['munResNome'] == 'Castanheiras'), ['munResNome','IDADEMAE','IDADEPAI']]
sinasc_3.dropna()
```

```python
sinasc_3.groupby(['munResNome']).agg([('média','mean'),'max','min'])
```

### 4. Qual o municipio que nasceu mais bebe no mês de março?
    - qual a quantidade de filhos vivos media, maxima, minima nesse municipio?
    - qual a idade media, maxima, minima dos pais nesse municipio?



```python
mes_nasc = []

data_nasc = list(sinasc_raw.DTNASC)

for i in data_nasc:
    x = i.split('-')
    mes_nasc.append(x[1])
    
sinasc_raw['mes_nasc'] = mes_nasc

sinasc_raw.mes_nasc
```

```python
sinasc_4 = sinasc_raw.loc[(sinasc_raw['mes_nasc'] == '03'), ['munResNome','QTDFILVIVO','IDADEPAI']]
sinasc_4
```

```python
sinasc_4['munResNome'].value_counts()
```

O município com o maior número de nascimentos no mês de março foi Porto Velho (744).

```python
sinasc_4.dropna().loc[(sinasc_4['munResNome'] == 'Porto Velho')].groupby(['munResNome']).agg([('média','mean'),'max','min'])
```
